# Pre Processing

```python
titanic.head()
```

```python
#One hot encoding --> conversion of categorical data into binary vectors
df = pd.get_dummies(titanic)
df.head()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support =0.1, use_colnames = True)
frequent_itemsets

#row 10 read as 11%of the total number of people(2201) belongs to class 2nd and is also an adult
```

```python
#relattionship between groups
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1.3)
rules
```
